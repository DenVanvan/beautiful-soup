```python
import pandas as pd
```

```python
import requests
from bs4 import BeautifulSoup
url = "http://pyclass.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/"
req = requests.get(url=url, headers=headers) # We use the headers
cont = req.text
```

```python
page_urls = ['']
soup = BeautifulSoup(cont, 'html.parser')
links=[i.get('href') for i in soup.find_all('a', class_='Page')][-2:]
page_urls += links
page_urls
```

```python
df = pd.DataFrame()
```

```python
for item in page_urls:
    url = "http://pyclass.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/"+item
    req = requests.get(url=url, headers=headers) # We use the headers
    cont = req.text
    soup = BeautifulSoup(cont, 'html.parser')

    divs = soup.find_all('div', class_ = 'propertyRow')
    for div in divs:
        price = div.find('h4', class_='propPrice').text.replace('\n', '').replace(' ', '')
        adress = div.find_all('div', class_='primaryDetails')[0].text.strip().replace('\n', ', ')
        
        area = None if div.find('span', class_='infoSqFt') == None else div.find('span', 
        class_='infoSqFt').text.replace(' Sq. Ft', '')
        beds = None if div.find('span', class_='infoBed') == None else div.find('span', 
        class_='infoBed').text.replace(' Beds', '')
        baths = None if div.find('span', class_='infoValueFullBath') == None else div.find('span', 
        class_='infoValueFullBath').text.replace(' Full Baths', '')
        
        for column_group in div.find_all('div', class_='columnGroup'):
            feature_group = column_group.find_all('span', class_='featureGroup')
            feature_name = column_group.find_all('span', class_='featureName')

            try:
                if 'Lot Size' in feature_group[0].text:
                    lot_size = float(feature_name[0].text.replace(',','').replace('Acre', '').replace('Under', '').replace('s', '').replace('1/2', '0.5').strip())
            except IndexError:
                lot_size = None

        df = df.append({'Address': adress, 'Price': price, 'Area': area, 'Beds': beds, 'Baths': baths, 'Lot Size': lot_size}, ignore_index=True)

df


```
